```python
# Importing Dependencies 
%matplotlib inline
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import auc
from matplotlib import style
from matplotlib import style
from sklearn.svm import SVC 

### Read the data
df = pd.read_csv("Household-living-cost-price-indexes-September-2022-quarter-time-series-indexes.csv")
df.head()
```

```python
# Checks Data Info
df.info
```

```python
# Looking for Unique values in the quarter column
df["quarter"].unique()
```

```python
# Split the quarter column into quarter and year
df[["year", "quarter"]] = df["quarter"].str.split('Q', n=1, expand=True)
df.head()
```

```python
# Count of rows
df.count()
```

```python
# Count of null values
df.isna().sum()
```

```python
# Drop Nulls then check for null values again
df = df.dropna()
df.isna().sum()
```

```python
# Display Data
df.head()
```

```python
# Renaming Columns 
df=df.rename(columns={"hlpi_name":"protocol id","nzhec_name":"expenditures","level":"groups","change.q":"quarterly","change.a":"annually"})
df.head()
```

```python
# Drop extra columns
df_clean=df.drop(['hlpi', 'nzhec_short', 'nzhec'], axis=1)
df_clean.head()
```

```python
# Change year to int
df_clean['year'] = df_clean['year'].astype('int')
```

```python
# Display data types
df_clean.dtypes
```

```python
# Filter year on 2019 and beyond
df_final = df_clean.loc[df_clean['year'] >= 2019]
df_final.head()
```

```python
# Recounting rows
rowcount = df_final.count()
rowcount
```

```python
# Adding a unique ID column to DataFrame
df_final['ID'] = np.arange(len(df_final))
df_final.head()
```

```python
# Put ID column at the front of my table
df_final = df_final[['ID', 'protocol id', 'series_ref', 'quarter', 'expenditures', 'groups', 'index', 'quarterly', 'annually', 'year']]
df_final.head()
```

```python
# Exporting the Clean Data
df_final.to_csv('Clean_Household_Data.csv', index=False)
```
